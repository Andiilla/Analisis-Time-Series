## Deret Curah Hujan Bogor

```python
#memanggil data curah hujan yang tersimpan dalam github
rr=pd.read_csv('https://raw.githubusercontent.com/lailynissa/tsapython/refs/heads/main/curahhujan_agg.csv')
```

```python
#menampilkan 10 data pertama
rr.head(10)
```

```python
#memeriksa tipe data
rr.dtypes
```

float64 adalah tipe data numerik dalam Python yang digunakan untuk menyimpan bilangan desimal (floating-point) dengan presisi 64-bit. Tipe ini biasanya digunakan untuk angka dengan nilai pecahan atau angka besar yang memerlukan presisi tinggi.

```python
#mengubah tipe data menjadi datetime
rr.tanggal=pd.to_datetime(rr.tanggal)
```

```python
#menampilkan plot deret waktu dari data
plt.figure(figsize=(10,5))
plt.plot(rr.tanggal,rr.rr,'-ok')
plt.xlabel('Waktu'), plt.ylabel('curah hujan (mm)')
plt.show()
```

Berdasarkan plot deret waktu yang dihasilkan apa yang dapat Anda interpretasikan?

1. Fluktuasi Curah Hujan: Data menunjukkan fluktuasi signifikan dalam curah hujan
(dalam satuan mm) dari waktu ke waktu antara tahun 2007 hingga 2013. Ada bulan-bulan dengan curah hujan tinggi yang diikuti oleh penurunan yang tajam.

2. Pola Musiman: Terlihat adanya pola tertentu yang berulang, yang mungkin menunjukkan pola musiman dalam curah hujan, meskipun tidak terlalu jelas.

3. Puncak Ekstrem: Beberapa periode menunjukkan lonjakan curah hujan yang sangat tinggi, misalnya pada awal tahun 2008 dan tahun 2011.

4. Periode Kekeringan: Terdapat juga periode di mana curah hujan sangat rendah, menunjukkan kemungkinan adanya kekeringan.

## Deret Harga Saham BRI

```python
bni.head(5)
```

```python
#menampilkan plot deret waktu
plt.figure(figsize=(10,5))
plt.plot(bni.Date,bni.Close,'-ok')
plt.xlabel('Waktu'), plt.ylabel('harga (Rp)')
plt.show()
```

Pola tersebut tidak stasioner dalam rataan karena pola tersebut tampak konstan atau tidak ada pola kecenderungan konstan karena ada kecenderungan naik/turun tapi dalam hal ini kecenderungan naik. Namun masih bisa tampak stasioner dalam ragam karena lebar atau jarak antar titik masih banyak yang berdekatan dan berulang.

Apa yang dapat Anda interpretasikan dari plot tersebut?

1. Pergerakan Harga:

  - Grafik menunjukkan perubahan harga dalam satuan Rupiah (Rp) terhadap waktu.
  - Pada awal periode (sekitar Oktober 2022), harga mengalami penurunan signifikan hingga mencapai titik terendah sekitar pertengahan Oktober 2022.
  - Setelah itu, harga mengalami tren kenaikan yang cukup tajam hingga awal November 2022.

2. Fluktuasi Harga:

  - Setelah mencapai puncak pada November 2022, harga relatif stabil tetapi masih menunjukkan sedikit fluktuasi hingga awal Desember.
  - Pada Desember 2022, terdapat kenaikan tajam yang diikuti penurunan.

3. Titik Penting:

  - Titik terendah: Pertengahan Oktober 2022 dengan harga sekitar Rp 8400.
  - Titik tertinggi: Awal Desember 2022 dengan harga mendekati Rp 9800.

4. Polanya:

  - Secara umum, pergerakan harga menunjukkan siklus naik-turun yang mungkin dipengaruhi oleh faktor musiman, ekonomi, atau lainnya.

## Deret Suhu Maksimum Jakarta

```python
#memanggil data suhu jakarta
suhu=pd.read_csv('https://raw.githubusercontent.com/lailynissa/tsapython/refs/heads/main/suhumax_jakarta.csv')
```

```python
#menampilkan 5 data pertama
suhu.head(5)
```

```python
#memeriksa type data
suhu.dtypes
```

```python
#mengubah type data
suhu.tanggal=pd.to_datetime(suhu.tanggal)
```

```python
#menampilkan plot deret waktu
plt.figure(figsize=(10,5))
plt.plot(suhu.tanggal,suhu.suhu,'-ok')
plt.xlabel('Waktu'), plt.ylabel('suhu (C)')
plt.show()
```

Apa interpretasi Anda terhadap plot deret waktu yang diperoleh?

```python
#menambahkan simbol bulan dalam plot deret waktu
mapping={1:'$J$',2:'$F$',3:'$M$',4:'$A$',
         5:'$M$',6:'$J$',7:'$J$',8:'$A$',
         9:'$S$',10:'$O$',11:'$N$',12:'$D$'}
plt.figure(figsize=(10,5))
suhu_new=suhu.reset_index()
plt.plot(suhu_new.tanggal,suhu_new.suhu, color='dimgray')
for i in range(len(suhu_new)):
    plt.plot(suhu_new.tanggal[i],suhu_new.suhu[i],
             marker=mapping[i%12+1],
             color='black', markersize=8)
plt.xlabel('Waktu'),plt.ylabel('suhu(C)')
plt.show()
```

Setiap 12 bulan sekali, terjadi tren musiman karena berulang setiap 12 bulan sekali pada bulan oktoober mengalami perulangan terkait kenaikan suhu

Setelah ditambahkan simbol bagaimana interpretasi Anda?
